```python
import yfinance as yf
import pandas as pd

def fetch_stock_data(ticker_symbol):
    # Get the data for the stock
    ticker_data = yf.Ticker(ticker_symbol)

    # Fetch the maximum available historical data
    historical_data = ticker_data.history(period="2y")

    # Normalize the datetime index of historical_data to remove the time component
    historical_data.index = historical_data.index.normalize()

    # Fetch earnings dates
    earnings_dates_data = ticker_data.get_earnings_dates(limit=30)
    
    # Normalize the datetime index of earnings_dates_data to remove the time component
    earnings_dates_data.index = earnings_dates_data.index.normalize()

    # Combine the historical data with earnings data
    combined_data = historical_data.merge(earnings_dates_data, left_index=True, right_index=True, how='left')

    return combined_data

# Fetch data for AAPL
apple_data = fetch_stock_data("GILD")
apple_data.tail(30)
```
